# FFT of Vowels (Harmonics, Timbre)

```python
import scikits.audiolab as audio
import scipy
from __future__ import division

def large_figure():
    fig = pylab.figure()
    fig.set_size_inches(14,8)
    
```

```python

(input_signal, sample_rate, bits) = audio.wavread("audio_files/vowel_ah.wav")
time_array = arange(0, len(input_signal)/sample_rate, 1/sample_rate)
```

```python
large_figure()
plot(time_array, input_signal)
```

```python
fft_out = numpy.fft.rfft(input_signal)
fft_mag = [sqrt(i.real**2 + i.imag**2)/len(fft_out) for i in fft_out]

large_figure()
plot(fft_mag[0:4000])
```
